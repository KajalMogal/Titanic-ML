# Data Cleaning
# We want to fill in missing age data instead of just dropping the missing age data rows. One way to do this is by filling in the mean age of all the   passengers (imputation). However we can be smarter about this and check the average age by passenger class. For example:

```python
plt.figure(figsize=(10, 7))
sns.boxplot(x='Pclass', y='Age', hue='Pclass', data=titanic,palette='winter')
```

# We can see the wealthier passengers in the higher classes tend to be older, which makes sense. We'll use these average age values to impute based on Pclass for Age.


```python
#titanic['Age'].mean()
titanic.groupby('Pclass')['Age'].mean()
```

```python
def impute_age(cols):
    Age = cols['Age']
    Pclass = cols['Pclass']

    if pd.isnull(Age):
        if Pclass == 1:
            return 38

        if Pclass == 2:
            return 29

        if Pclass == 3:
            return 25

    else :
            return Age
```

```python
titanic['Age']=titanic[['Age', 'Pclass']].apply(impute_age, axis=1)
```

# Converting categotical features
# We'll need to convert categorical features to dummy variables using pandas! Otherwise our machine learning algorithm won't be able to directly take in those features as inputs.

```python
sex=pd.get_dummies(titanic['Sex'], drop_first=True, dtype=int)
embarked=pd.get_dummies(titanic['Embarked'], drop_first=True, dtype=int)
```

```python
titanic.drop(['Sex', 'Embarked', 'Name', 'Ticket'], axis=0, inplace=True)
titanic.head()
```

```python
titanic=pd.concat([titanic, sex, embarked], axis=1)
```

```python
titanic.head()
```

```python
titanic.drop(['Name', 'Ticket', 'Embarked', 'Sex'], axis=1, inplace=True)
```

```python
titanic.head()
```

# splitting our data into a training set and test set (there is another test.csv file that you can play around with in case you want to use all this data for training).

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(titanic.drop('Survived', axis=1), titanic['Survived'], test_size=0.30, random_state=101)
```

```python
titanic.head()
```

```python
print(X_train.shape)
print(X_test.shape)
```

# Training And Predictions

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression(max_iter=1000)
logmodel.fit(X_train, y_train)
```

```python
predictions=logmodel.predict(X_test)
```

# Evaluation

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```
